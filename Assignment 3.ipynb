```python
print("====== SIMPLE CALCULATOR =======")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. Power")
choice = int(input("Choose Option: "))
arr = []
if choice == 1:
    x = int(input("How many numbers do you want to add: "))
    total = 0
    for i in range(x):
        y = int(input("Enter number: "))
        arr.append(y)
        total += y
    print("Numbers:", arr)
    print("Sum:", total)
elif choice == 2:
    x = int(input("How many numbers do you want to subtract: "))
    for i in range(x):
        y = int(input("Enter number: "))
        arr.append(y)
    ans = arr[0]
    for i in range(1, len(arr)):
        ans = ans - arr[i]
    print("Numbers:", arr)
    print("Subtraction Result:", ans)
elif choice == 3:
    x = int(input("How many numbers do you want to multiply: "))
    ans = 1
    for i in range(x):
        y = int(input("Enter number: "))
        arr.append(y)
        ans = ans * y
    print("Numbers:", arr)
    print("Multiplication Result:", ans)
elif choice == 4:
    x = int(input("How many numbers do you want to divide: "))
    for i in range(x):
        y = int(input("Enter number: "))
        arr.append(y)
    ans = arr[0]
    for i in range(1, len(arr)):
        if arr[i] != 0:
            ans = ans / arr[i]
        else:
            print("Error: Division by zero is not allowed.")
            ans = "Undefined"
            break
    print("Numbers:", arr)
    print("Division Result:", ans)
elif choice == 5:
    base = int(input("Enter base number: "))
    exponent = int(input("Enter exponent: "))
    ans = base ** exponent
    print("Power Result:", ans)
else:
    print("Invalid choice! Please enter between 1 and 5.")
```

```python
numbers = [1, 2, 3, 2, 4, 5, 1]
appeared = []
duplicates = []
for num in numbers:
    if num in appeared and num not in duplicates:
        duplicates.append(num)
    else:
        appeared.append(num)
print("Duplicates:", duplicates)
```
