```python
maths=float(input("Maths obtained Marks out of 100: "))
physics=float(input("Physics obtained Marks out of 100: "))
Computer=float(input("Computer obtained Marks out of 100: "))
Islamiath=float(input("Islamiath obtained Marks out of 100: "))
English=float(input("English obtained Marks out of 100: "))
```

```python
sum= maths+physics+Computer+Islamiath+English
per=(sum/500)*100
roundoff=round(per,2)
```

```python
print("PERCENTAGE: ", roundoff)
if per>=80: print("GRADE: A+")
elif per>=70 and per<80: print("GRADE: A")
elif per>=60 and per<70: print("GRADE: B")
elif per>=50 and per<60: print("GRADE: C")
elif per<50: print("GRADE: F")
if per>=50: print("Passed!.....")
else: print("Failed!....")
```
