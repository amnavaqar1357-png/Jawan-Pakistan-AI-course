```python
x=int(input("Enter 1st number: "))
y=int(input("Enter 2nd Number: "))
z=input("Enter operator: ")
```

```python
if z=="+":
  print(x," + ",y," = ",(x+y))
elif z=="-":
  print(x," - ",y," = ",(x-y))
elif z=="*":
  print(x," * ",y," = ",(x*y))
elif z=="/":
  if(y==0): print("Division is not possible")
  print(x," - ",y," = ",(x/y))
elif z=="%":
  if(y==0): print("Division is not possible")
  print(x," % ",y," = ",(x%y))
else:print("Invalid Operator")
```
