```python
maths=float(input("Maths obtained Marks out of 100: "))
physics=float(input("Physics obtained Marks out of 100: "))
Computer=float(input("Computer obtained Marks out of 100: "))
Islamiat=float(input("Islamiath obtained Marks out of 100: "))
English=float(input("English obtained Marks out of 100: "))
```

```python
total = maths + physics + Computer + Islamiat + English
percentage = (total / 500) * 100
percentage_rounded = round(percentage, 2)
```

```python
if percentage >= 80:
    grade = "A+"
elif percentage >= 70:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 50:
    grade = "C"
else:
    grade = "F"
```

```python
status = "Passed!...." if percentage >= 50 else "Failed!...."
print("\n==== MARKSHEET ====")
print("Maths:      ", maths)
print("Physics:    ", physics)
print("Computer:   ", Computer)
print("Islamiat:   ", Islamiat)
print("English:    ", English)
print("=====================")
print("Total Marks:", total, "/ 500.0")
print("Percentage: ", percentage_rounded, "%")
print("Grade:      ", grade)
print("Status:     ", status)
print("====================")
```
